# 라이브러리

```python
import pandas as pd

import matplotlib.pyplot as plt
plt.rc("font", family="Malgun Gothic")  # 한글 맑은고딕 폰트 사용
```

# 종관기상관측(ASOS) 데이터 불러오기

```python
df = pd.read_csv("../data/OBS_ASOS_MI_159_20220906.csv", encoding="cp949")
df
```

## 컬럼 데이터타입

```python
df.dtypes
```

## 일시 컬럼을 datetime 타입으로 변환

```python
df["일시"] = pd.to_datetime(df["일시"])
df["일시"]
```

```python
df.dtypes
```

## 일시 컬럼을 인덱스화

```python
df.index = df["일시"]
df.drop("일시", axis="columns", inplace=True)
df
```

# QC 플래그

플래그 종류(의미): 0(정상), 1(오류), 9(결측)

```python
qc_columns = ["기온", "풍향", "풍속", "현지기압", "해면기압", "습도"]
```

```python
df["기온 QC플래그"].unique()
```

```python
df["풍향 QC플래그"].unique()
```

```python
for column in qc_columns:
    print(column, "QC플래그 종류:", df[column+" QC플래그"].unique())
```

만약, QC 플래그가 1 또는 9인 경우 해당되는 변수값을 NaN (Not a Number)으로 대체해야 함

```python
nan_temperature = df["기온 QC플래그"].isin([1, 9])
df.loc[nan_temperature, "기온(°C)"] = pd.NA
```

# 풍속 그래프

```python
df["풍속(m/s)"]
```

```python
plot = df["풍속(m/s)"].plot() # 풍속 그래프를 plot 객체에 저장
plt.show()
```

## X/Y 축

### X축 이름

```python
plot.set_xlabel("시간(시:분)")
```

### Y축 이름

```python
plot.set_ylabel("풍속(m/s)")
```

```python
plot = df["풍속(m/s)"].plot()
plot.set_xlabel("시간(시:분)")
plot.set_ylabel("풍속(m/s)")
plt.show()
```

```python
plot = df["풍속(m/s)"].plot()
plot.set_xlabel("시간(시:분)", fontsize=12)
plot.set_ylabel("풍\n속\n(m/s)", fontsize=12, rotation=0, labelpad=20)  # \n은 엔터(줄바꿈)
plt.show()
```

## 선 스타일(색, 두께)

```python
plot = df["풍속(m/s)"].plot(c="k", lw=0.5)  # c (color), lw (linewidth)
plot.set_xlabel("시간(시:분)", fontsize=12)
plot.set_ylabel("풍\n속\n(m/s)", fontsize=12, rotation=0, labelpad=20)  # \n은 엔터(줄바꿈)
plt.show()
```

## 그림 사이즈

```python
plot = df["풍속(m/s)"].plot(c="k", lw=0.5, figsize=(5, 3))  # figsize=(가로, 세로)
plot.set_xlabel("시간(시:분)", fontsize=11)
plot.set_ylabel("풍\n속\n(m/s)", fontsize=11, rotation=0, labelpad=20)
plt.show()
```

# 기압 그래프 추가

```python
df["현지기압(hPa)"]
```

```python
df["현지기압(hPa)"].plot()
plt.show()
```

```python
plot = df["풍속(m/s)"].plot(c="k", lw=0.5, figsize=(5, 3))
plot.set_xlabel("시간(시:분)", fontsize=11)
plot.set_ylabel("풍\n속\n(m/s)", fontsize=11, rotation=0, va="center", labelpad=20)
# 기압 그래프 추가
twin_plot = plot.twinx()
df["현지기압(hPa)"].plot(c="k", lw=0.5, ax=twin_plot)
twin_plot.set_ylabel("기\n압\n(hPa)", fontsize=11, rotation=0, va="center", labelpad=20)
plt.show()
```
